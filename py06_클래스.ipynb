##### 1. 클래스 정의하기

```python
# 1. 클래스 정의하기

class Person:
    def __init__(self, name, age, addr): #초기화 변수들을 다 넣는다...자동으로 호출
        self.name = name
        self.age = age
        self.addr = addr
    def greeting(self): #Person 클래스의 메소드이다...란 의미(self)
        print('안녕하세요')
        
    def sing(self):
        self.greeting()
        print('랄랄랄 루루루')
        
```

##### 2. 클래스 객체 생성하기

```python
#james = Person()
james = Person('James Corner',33,'NY')
peter = Person('PETER',44,'Brandon')
jane = Person('JANE',55,'Texas')
```

##### 3. 멤버 접근해서 정보 확인하기

```python
james.name
james.age
james.addr
james.greeting()
james.sing()
```
