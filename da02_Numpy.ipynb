### 랜덤함수와 seed값 설정하기
    seed값은 한번만 설정해주면 계속 적용된다.
    시드는 보통 현재시각등을 시용해서 자동으로 정해지지만
    우리가 수동으로 설정할 수 있다.
    특정한 시드값이 사용되면 그 다음에 만들어지는 난수들을 예측할 수 있다.
    

```python
import numpy as np
```

```python
np.random.seed(0)

arr1 = np.random.randn(4,4)
arr1
```

```python
arr2 = np.arange(1,10)
arr2

arr3 = np.arange(1,10).reshape(3,3)
arr3
```

### 인덱싱, 슬라이싱

```python
narray = np.arange(5)
narray
```

```python
# 1. 인덱싱, 슬라이싱으로 값 추출하기
print(narray[1],narray[4]) # 또는 narray[1],narray[-1]
print(narray[1:])
print(narray[:])
print(narray[2:4])
```

```python
narray2 = np.random.randint(10,20,16).reshape(4,4)
narray2
```

```python
# 1) narray2 중에서 첫번재 행의 세번째 값 12를 가져오려면?
narray2[0][2]
narray2[0,2]

# 2) 2번째 라인 [10,14,15,15] 전체를 출력
narray2[1]
narray2[1:2, 0:4] #2차원 출력
narray2[1:2, :] #2차원 출력
narray2[1, :] #1차원 출력

# 3) 전체 행에 대해서 4번째 열 [10,15,11,11] 출력 
narray2[0:4, 3:4]
narray2[:, 3:]
narray2[:, 3]

# 4) 전체 행에 대해서 1,2번째까지의 열을 출력
narray2[:,:2]
```

### 인덱싱과 조건 슬라이싱 

```python
arr1
#1. 논리연산자 사용해서 값을 받아오거나 값을 변경

arr1[arr1>0]
arr1[arr1<0]
arr1[arr1<0] = 0
#원본도 함께 바뀐다.
arr1
```

```python
#2. where함수 
arr1_1 = np.where(arr1>0,arr1,-1)
arr1_1
```

```python
#3. 1,2를 가지고 응용하기
arr2 = np.array([[1,2],[3,4],[5,6]])
arr2

#2보다 큰 수만 출력하기
arr2[arr2>2]
```

```python
bool_idx = arr2>2
print(bool_idx)

print(arr2[bool_idx])
```
